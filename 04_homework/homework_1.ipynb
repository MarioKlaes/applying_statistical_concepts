```python
# Import standard libraries
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Import specific objects
from textwrap import wrap # to avoiding label overlapping in plots
from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)

# Load dataset
Auto = load_data('Auto')
Auto
```

_a)_	Use the `sm.OLS()` function to perform a simple linear regression with `mpg` as the response and `horsepower` as the predictor. Use the `model.summary()` function to print the results. Comment on the output

(i) Is there a relationship between the predictor and the response?
- Yes.

(ii) How strong is the relationship between the predictor and the response?
- R-squared (uncentered):	0.649
- 64.9% of the MPG are explained by Horse Power

(iii) Is the relationship between the predictor and the response positive or negative?
- the coeficent is 0.1788. Since it is a positive numberm the relatioship is posaitive.

(iv) What is the predicted mpg associated with a horsepower of 98? What are the associated 95 % confidence and prediction intervals?
- mpg_predicted = regresion_model.predict (98)
- y | mpg_predicted = 17.52630402


```python
#Add your code here
mpg = Auto['mpg'] # miles per galon
horsepower = Auto['horsepower']

horsepower_reshaped = horsepower.values.reshape(-1, 1)

# It sets up the linear regression equation to predict the target variable (y) based on the independent variables (X).
model = sm.OLS(mpg, horsepower_reshaped) # to specify the Ordinary Least Squares (OLS) regression model

regresion_model = model.fit() # to fit the model
print (regresion_model.summary())

mpg_predicted = regresion_model.predict (98)
print (mpg_predicted)

#(iv) What is the predicted mpg associated with a horsepower of 98? 
# What are the associated 95 % confidence and prediction intervals?
```

_b)_	Plot the response and the predictor in a new set of axes `ax` using `scatter`. Use the `ax.plot()` function with the predictor and the model prediction as defined in the lab to display the least squares regression line.

```python
#Add your code here
# Plot 'medv' against 'rm'
plt.scatter(horsepower, mpg, label='Data')

plt.xlabel('Horse Power')
plt.ylabel('MPG')

# Plot the regression line
plt.plot(horsepower_reshaped, regresion_model.predict(horsepower_reshaped), color='red', linewidth=2, label='Regression Line')

plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

mpg = Auto['mpg'] # Y miles per galon
horsepower = Auto['horsepower'] # X

# Reshape 'rm' to a 2D array for LinearRegression
#  This line reshapes the feature variable rm into a 2D array (matrix) with a single column. 
#  It’s necessary for using it in the LinearRegression model.
horsepower_reshaped = horsepower.values.reshape(-1, 1)

# Fit a linear regression model

lrm_horsepower_mpg = LinearRegression()
lrm_horsepower_mpg.fit(horsepower_reshaped, mpg)


# Plot 'medv' against 'rm'
plt.scatter(horsepower, mpg, label='Data')
plt.xlabel('horsepower')
plt.ylabel('mpg')

# Plot the regression line
plt.plot(horsepower_reshaped, lrm_horsepower_mpg.predict(horsepower_reshaped), color='red', linewidth=2, label='Regression Line')

plt.legend()
plt.show()
```
