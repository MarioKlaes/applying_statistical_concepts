## Prediction vs Inference ##

![image](./prediction_vs_inference.png)

# Predictins #

- Find the 'y' based on x --> f(x) + e = y

The prediction is affected by the error 'e'.
To evaluate the quality of the prediction, it is necessary use the *p-value*

# Inference #

- Answer the question of how much X influences Y.
On the example, how strongly TV affect sales? 
Who affect more sales (or who generates more sales per dolar invested): TV, Radio or News Papaer?

# Training Data vs Testing Data or 80/20 #

- From the data ser, separate 80% for training the model to create the prediction function (f(x) + e = y)
- Then get 20% of the sample to test the prediction funcion

# Overfitting #

When the prediction model is too precise it may fail to predict future observations reliably.
In other words, the model gets supper-specilized in the data set.

# Supervised Learning vs Unsupervised Learining #

Supervised learning: use model to predict a response based on the predictor f(x) + e = y

Unsupervised learning: used to investigate the relationship bewteen variables or observations
i.e. clustering


# Qialitative vs Quantitative Variables #

- Quantitative reposnses ==> Regression Problems
- Qualitative responses ==> Classification Problems


# Assessing Model Acuracy #

- For Regression: use Mean Squared Error (MSE)
- smaller MSE the better
